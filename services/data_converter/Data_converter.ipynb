## Imports

```python
from scipy.io import loadmat
import pandas as pd 
```

## Import data and get overview

```python
mat_file_path = "./cars_train_annos.mat"
raw_data = loadmat(mat_file_path)
raw_data
```

```python
annotations = raw_data["annotations"]
annotations
```

The imported data is structured as array with a single row 
row = many different tuples. Each tuple has tuples with array
access the needed data via slicing & list compreheinsion

## Save the needed data into lists 

```python
# for fname
image_names = [i[5][0] for i in annotations[0]]
image_names
```

```python
# for the int values one more dimension of slicing (see data structure)
bbox_x1 = [i[0][0][0] for i in annotations[0]]
bbox_x1
```

```python
# same way for the other int values 
bbox_y1 = [i[1][0][0] for i in annotations[0]]
bbox_x2 = [i[2][0][0] for i in annotations[0]]
bbox_y2 = [i[3][0][0] for i in annotations[0]]
class_ids = [i[4][0][0] for i in annotations[0]]

# bbox_y1
# bbox_x2
# bbox_y2
# class_ids
```

## Package the extracted data into pandas dataframe 

```python
df_struct = {"img_name": image_names,
             "x1": bbox_x1, 
             "x2": bbox_x2,
             "y1": bbox_y1, 
             "y2": bbox_y2, 
             "class": class_ids}

df = pd.DataFrame(df_struct)
df
```

## Import meta data (classes) and getting overview

```python
meta_data_path = "./cars_meta.mat"
meta_data = loadmat(meta_data_path)
meta_data
```

```python
classes = meta_data["class_names"]
class_names = [i[0] for i in classes[0]]
class_names
```

## package class name into a pandas dataframe 

```python
classes_df = pd.DataFrame({"class": class_names})
classes_df.head(200)
```

## Generate the final Dataframe and export to csv 

```python

df['class-name'] = df['class'].map(classes_df['class'])

# there are somehow class ids which are not available (ID 196)
# it's not because indexing, id=1 correct name, id=4 coorect name and so on 
# just 40 records with 196. They will drop 
df

df.isna().sum() # 40 in class-name
df[df["class-name"].isna()] # every record has class 196
```

```python
# drop
final_df = df.dropna().drop("class", axis=1)
final_df
```

```python
target_path = "./train_data.csv"
final_df.to_csv(target_path, index=False)
```
