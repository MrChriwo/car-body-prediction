```python
data = cleanset[cleanset["cleanlab_clean_label"] != "None"]
data
```

```python
# Load the DataFrame
df = data

# Iterate over each row in the DataFrame
for index, row in df.iterrows():
    file_path = row['id']
    new_folder = row['cleanlab_clean_label']
    try: 
        # Check if the file exists
        if os.path.exists(file_path):
            # Create the new folder if it doesn't exist
            if not os.path.exists(new_folder):
                os.makedirs(new_folder)

            # Move the file to the new folder
            shutil.move(file_path, new_folder)
            print(f"Moved file {file_path} to {new_folder}")
        else:
            print(f"File {file_path} does not exist.")
    except: 
        continue

print("File moving completed.")
```

```python
# clean and save dataframe 
files = listdir("../data/images/")
df = pd.read_csv("../data/meta/final_dataset_raw.csv")

for index, data in df.iterrows(): 
    if data["name"] not in files: 
        df.drop(index, axis=0, inplace=True)

df.to_csv("../data/meta/final_dataset_raw.csv")
```
